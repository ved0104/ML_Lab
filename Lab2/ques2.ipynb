```python
for col in df:
    if df[col].values==["ENTRY_DATE","EVENT_LCL_DATE","EVENT_LCL_TIME"]:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    if df[col].values==["ACFT_MAKE_NAME","LOC_STATE_NAME","ACFT_MODEL_NAME","RMK_TEXT","FLT_PHASE","EVENT_TYPE_DESC","FATAL_FLAG"]
    if df[col].dtype == 'object':
        df[col].fillna('Unknown', inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
df.shape
```

```python
df.sample(10)
```

```python
def encode_categorical_from_scratch(df):
    df = df.copy()
    for col in df.columns:
        if df[col].dtype == 'object':
            unique_vals = df[col].dropna().unique()
            # Create a mapping from category to integer
            mapping = {val: idx for idx, val in enumerate(unique_vals)}
            # Map the column values to integers, keep NaNs as is
            df[col] = df[col].map(mapping)
    return df
```

```python
df_encoded = encode_categorical_from_scratch(df)
```

```python
df_encoded.sample(5)
```

```python
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm')
plt.figure(figsize=(12,8))
plt.title('Correlation Heatmap')
```
